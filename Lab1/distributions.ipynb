```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.preprocessing as skp
import matplotlib.pyplot as plt
```

```python
import scipy.stats as sts
```

```python
data = pd.read_csv("sample.csv")
```

```python
data.columns
```

```python
sns.histplot(data=data, x = "State_")
```

```python
data14, data56 = data[data['State_']<=4] , data[data['State_']>4]
```

```python
def plot_histogram(featurename):
    _, ax = plt.subplots(1,2, width_ratios=[10,10])
    sns.histplot(data=data14, x = featurename, ax = ax[0])
    sns.histplot(data=data56, x = featurename, ax = ax[1])
```

```python
plot_histogram('Ta_mean')
```

```python
sts.shapiro(data14['Ta_mean'].apply(lambda x: np.sqrt(data14['Ta_mean'].max() - x )))
```

```python
sts.probplot(data['Ta_mean'])
```

```python
for feat in ['Ta_mean', 'Tsurf_mean',
       'Water_mean', 'Speed_mean', 'Height_mean', 'Tdew_mean',
       'Friction_mean',
       'Direction_min', 'Direction_max', 'maxtempC', 'mintempC', 'totalSnow_cm',
       'sunHour', 'uvIndex', 'DewPointC', 'FeelsLikeC', 'HeatIndexC',
       'WindChillC', 'WindGustKmph', 'cloudcover', 'humidity', 'precipMM',
       'pressure', 'tempC', 'visibility', 'winddirDegree', 'windspeedKmph']:
    print(feat)
    plot_histogram(featurename= feat)
```
